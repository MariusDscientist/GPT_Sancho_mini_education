```python
import numpy as np

x = np.array([1.0, 0.0, -3.0, 4.0])

gamma = np.array([2.0,1.0, 0.5,3.0])
beta = np.array([0.1,0.2,0.3,0.4])
```

```python
media = np.mean(x)
print(media)
```

```python
epsilon = 1e-5

var = np.var(x)
print(var)
```

```python
numerador = x-media

denominador = np.sqrt(var + epsilon)

vector_norm = numerador / denominador

vector_norm
```

```python
vector_final = vector_norm * gamma + beta

vector_final
```
